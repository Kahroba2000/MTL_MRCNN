```python
import configparser
import os
import dataset_splitter

# This file first downloads images and store them in the directory that is defined in the config_default.py.
# It then changes the json file to point to the local or google colab copy
# It then calls dataset_splitter, which randomly splits the data into training, validation, and test sets
# based on split values defined in config_default.py


def configure_json():
    print("[INFO] Configuring JSON file and downloading images.")

    config = configparser.ConfigParser()
    config.read('config_default.ini')
    
    json_path = config.get('FILE_PATHS','json_path')
    print (json_path)
    import json
    import urllib.request

    with open(json_path) as f:
        dataset = json.load(f)
        for i in dataset['images']:
            # downloading images to local copy
            url = (i["file_name"])
            print("Downloading:", url)
            image_dir = config.get('FILE_PATHS', 'json_image_target_dir_local')
            os.makedirs(image_dir, exist_ok=True)
            image_filename = os.path.basename(url)
            image_filename = image_filename[image_filename.rindex('-') + 1:]
            image_path = os.path.join(config.get('FILE_PATHS', 'json_image_target_dir_local'), image_filename)
            print("Local storage location:", image_path)
            urllib.request.urlretrieve(url, image_path)

            # changing json to point to local or google drive copy of images
            # location is defined by a config value which is automatically changed depending on a tick box
            file_name = config.get('FILE_PATHS', 'json_image_target_dir_current') + "/" + image_filename
            print("Image reference path:", file_name)
            (i["file_name"]) = file_name

    print("[INFO] JSON configured and files downloaded successfully.")
    dataset_splitter.split_data(dataset)
    
    
configure_json()
```
